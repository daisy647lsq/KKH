```r
library(ggplot2)
balance <- read.csv("balance.csv")
df_balance<-data.frame(balance)
```

```r
head(df_balance)
```

```r
p<-ggplot(df_balance,
       aes(x=After,
           y=smd,
           color=matching
           ))+
  geom_point()
p1<-p+scale_x_continuous("Before Matching                     After Matching",limits = c(-0.5, 1.5)) +
  scale_y_continuous("SMD")+ggtitle("Standard Mean Difference Before & After Matching with PS", subtitle = NULL)+theme(
      axis.ticks.x=element_blank(),axis.text.x=element_blank(),axis.title.x=element_blank())
p1
```

```r
cutoff <- data.frame( x = c(-Inf, Inf), y = 0.05, cutoff = factor(0.05) )
ggplot(df_balance,
       aes(x=x,
           y=p_value,
           color=matching))+geom_point()+geom_line(aes( x, y, linetype = cutoff),color="black",cutoff)+
 scale_y_continuous("P-value")+ggtitle("P-value Before & After Matching with Propensity Score", subtitle = NULL)+theme(
      axis.ticks.x=element_blank(),axis.text.x=element_blank(),axis.title.x=element_blank()) 
```
